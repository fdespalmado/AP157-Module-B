```python
from google.colab import drive
drive.mount('/content/drive')

import matplotlib.pyplot as plt
import numpy as np
import skimage.io

# read target phrase image
file_path = "/content/drive/My Drive/S2-2021 APPHY157/B/3/"
fname = "RAIN.bmp"
Img = skimage.io.imread(file_path+fname)
plt.figure(1)
plt.imshow(Img, cmap='gray')

# read A image
aname="A.bmp"
A = skimage.io.imread(file_path+aname)
plt.figure(2)
plt.imshow(A, cmap='gray')

# read I image
iname="I.bmp"
I = skimage.io.imread(file_path+iname)
plt.figure(3)
plt.imshow(I, cmap='gray')
```

```python
from skimage import data
from skimage.color import rgb2gray
import cv2 as cv

def correlation(template):  
  FImg = np.fft.fft2(Img)
  AFFT = np.fft.fft2(template)  
  P = np.conj(FImg)*AFFT    

  smallp = np.fft.fft2(P)
  FImage = rgb2gray(np.fft.fftshift(abs(smallp)))
  plt.imshow(FImage, cmap='gray')
  
  img = FImage/ FImage.max()
  img = img* 255
  #cv.imwrite(f"{file_path}correlation{str(template)}.bmp", img)
```
